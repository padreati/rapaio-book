```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
tp = np.random.normal(0.15, 0.15, 200)
fp = np.random.normal(0.6, 0.15, 200)
```

```python
plt.rcParams['figure.figsize'] = (16,10)
plt.hist(tp, bins=30, color='r', alpha=0.5)
plt.hist(fp, bins=30, color='b', alpha=0.5)
plt.xlim(0, 1)
plt.vlines([0.5], 0, 25)
plt.show()
```

```python
ctp = np.cumsum(tp)
cfp = np.cumsum(fp)

plt.plot(ctp)
plt.plot(cfp)
plt.show()
```
