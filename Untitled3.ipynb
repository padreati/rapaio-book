```java
int N = 30;
int K = 4;
```

```java
1 - 10.0/16
```

```java
Binomial bin = Binomial.of(10.0/16, N)
```

```java
// probability for less than 4
bin.cdf(K-1)
```

```java
// probability for greater than
1-bin.cdf(N-K)
```

```java
double q = bin.cdf(K-1) + 1 - bin.cdf(N-K)
```

```java
double mean = q*1_000_000
```

```java
mean
```

```java
// standard deviation
double sd = sqrt(1_000_000 * q * (1-q))
```

```java
// 2 standard deviation interval
WS.println("[" + (mean-2*sd) + "," + (mean+2*sd) + "]")
```

```java
VarDouble x = VarDouble.seq(0, N, 1)
```

```java
VarDouble y = VarDouble.from(x, bin::pdf)
```

```java
WS.image(lines(x,y))
```
