```java
int N = 100_000;
double p = 0.015;
```

```java
VarInt ns = VarInt.seq(1,N);
VarDouble qs = VarDouble.empty();
for(int n : ns) {
    qs.addDouble(Binomial.of(p,n).mean());
}
```

```java
// agrousti-coull
double z = Normal.std().quantile(0.999);
```

```java
double ubound(double z, double k, double n) {
    double phat = (k+z*z/2) / (n+z*z);
    return phat + z * Math.sqrt((phat*(1-phat))/(n+z*z));
};
double lbound(double z, double k, double n) {
    double phat = (k+z*z/2) / (n+z*z);
    return phat - z * Math.sqrt((phat*(1-phat))/(n+z*z));
};
```

```java
VarDouble ubound = VarDouble.from(N, row -> ubound(z,p*ns.getDouble(row),ns.getDouble(row))*ns.getDouble(row));
VarDouble lbound = VarDouble.from(N, row -> lbound(z,p*ns.getDouble(row),ns.getDouble(row))*ns.getDouble(row));
```

```java
WS.image(
    lines(ns, ubound,color('b'))
    .lines(ns, qs, color('r'))
    .lines(ns, lbound, color('b'))
)
```
